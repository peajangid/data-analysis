```python
# importing important lybraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the data from:
df=pd.read_csv(r'C:\Users\dell\Desktop\python and r\Python_Diwali_Sales_Analysis\Python_Diwali_Sales_Analysis\Diwali_Sales_Data.csv',encoding='unicode_escape')
               
```

```python
#getting all the information of the dataframe for understanding the content of the data:
df.shape
df.columns
df.describe()
df.info()
```

```python
# checking the  the null values:
pd.isnull(df)
```

```python
#couting the total null values in the particular column
pd.isnull(df).sum()
```

```python
#dropping the empty columns:
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#dropping the null values of the dataframe and change the dataframe:
df.dropna(inplace=True)
df.info()
```

```python
#changing the type of the column Amount:
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

Exploratory Data Analysis

```python
#plotting a barplot of the Gender Vs count. This gives the info about the purchases made by the female and the male seperately
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(3,3)})

```

```python
# plotting the graph of Gender Vs Amount 
Sales_set=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_set ,x='Gender',y='Amount')
sns.set(rc={'figure.figsize':(10,10)})

```

```python
# plotting thr graph of Age Group Vs their count of purcases accoring to gender
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
# the graph of the Age group Vs the Amount
Sales_Age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_Age,x='Age Group',y='Amount')
```

```python
# the Graph of the State Vs orders 
Sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(8)
sns.barplot(data=Sales_state,x='State',y='Orders')
sns.set(rc={'figure.figsize':(15,5)})
```

```python
#Graph of State Vs Amount
Sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
sns.barplot(data=Sales_state,x='State',y='Amount')
```

```python
#Graph of the Marital Status Vs Count of the purchase
ax=sns.countplot(data=df,x= 'Marital_Status',hue='Gender')
sns.set(rc={'figure.figsize':(5,5)})
```

```python
# Graph of the Marital Status Vs Amount
Sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
#Graph of the Occupation Vs Count
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# The graph of Occupation Vs The Amount
Sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_state,x='Occupation',y='Amount')
```

```python
# graph of the Product category Vs Count
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Graph of the Product category Vs Amount
Sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
sns.barplot(data=Sales_state,x='Product_Category',y='Amount')
```

```python
# Graph of the Product Id Vs the Amount 
Sales_state=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
sns.barplot(data=Sales_state,x='Product_ID',y='Amount')


```
